```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
from sympy.ntheory import factorint

t = 2000
d = 7242624
results = [sum(v * n for v, n in factorint(i).items()) for i in range(d, d + t)]
```

```python
from functools import reduce

y_data = [np.inf]
for i in range(len(results)):
    y_data.append(min(y_data[-1], results[i]))
del y_data[0]

print(y_data)
n_fill = np.argmin(results)
print(n_fill+d)
print(factorint(n_fill+d))

plt.plot(y_data)
```
